PyTorch Linear Regression

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
#creating a function f(x) with a slope of -5
X=torch.arange(-5,5,0.2).view(-1,1)
# you go from -5 to 5 with jump of 0.2 -.48 -.44
# print(X)
func=-5*X
#function is our ground truth , X takes a value
```

```python
#convert the torch array into numpy array. internally numpy is prefered
plt.plot(X.numpy(),func.numpy(),'r',label='func')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid('True',color='y')
plt.show()
```

```python
X.size()
```

```python
Y=func+1.7 * torch.randn(X.size())
#we are introducting error of 1.7
```

```python
#convert the torch array into numpy array. internally numpy is prefered
plt.plot(X.numpy(),Y.numpy(),'b+',label='Y')
plt.plot(X.numpy(),func.numpy(),'r',label='func')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid('True',color='y')
plt.show()
```

```python
#blue points are real data we want to show the blue dots to the machine and help it predict the red line
```

```python
#convert the torch array into numpy array. internally numpy is prefered
plt.plot(X.numpy(),Y.numpy(),'b+',label='Y')
# plt.plot(X.numpy(),func.numpy(),'r',label='func')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid('True',color='y')
plt.show()
```

```python
#hypo w replaces teeta with w
def forward(x):
    return w*x
```

```python
#need a loss function / cost/ energy Mean square Error

def criterion(y_pred,y):
  return torch.mean((y_pred-y)**2)
```

```python
w=torch.tensor(-10.0,requires_grad=True)
#random guess with 1 dimension with random value of -10
```

```python
step_size=0.1 #alpha learning rate
loss_list=[]
iter=20
```

```python
for i in range(iter):
  y_pred=forward(X)
  loss=criterion(y_pred,Y)
  loss_list.append(loss.item())
  loss.backward() #all the backpropagtion derivatives

  w.data=w.data-step_size*w.grad.data #updation of w
  w.grad.data.zero_() #zeroing the graients after each iteration important
  print('{},\t{},\t{}'.format(i,loss.item(),w.item()))
```

```python
#now visulize the loss
plt.plot(loss_list,'r')
plt.tight_layout()
plt.grid('True',color='y')
plt.xlabel('Epochs/Iterations')
plt.ylabel('Loss')
plt.show()
```
